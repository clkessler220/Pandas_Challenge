```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Create data for the district summary dataframe, use information from above
total_number_of_schools = len(school_data_complete["school_name"].unique())
total_number_of_students = len(school_data_complete["Student ID"])
total_budgets = school_data["budget"].sum()
average_math_score = student_data["math_score"].mean()
average_reading_score = student_data["reading_score"].mean()
math_scores_70_df = school_data_complete.loc[(school_data_complete["math_score"] >= 70)]
math_scores_70_pct = (math_scores_70_df["math_score"].count()/total_number_of_students)*100
reading_scores_70_df = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]
reading_scores_70_pct = (reading_scores_70_df["reading_score"].count()/total_number_of_students)*100
overall_passing = (reading_scores_70_pct + math_scores_70_pct)/2

# Create a list of dictionaries
district_summary = [{"Total Schools": total_number_of_schools, "Total Students": total_number_of_students, 
            "Total Budget": total_budgets, "Average Math Score": average_math_score,
            "Average Reading Score": average_reading_score, "% Passing Math": math_scores_70_pct,
            "% Passing Reading": reading_scores_70_pct, "Overall Passing Rate": overall_passing}]

district_summary_df = pd.DataFrame(district_summary)

district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# created a new dataframe with data built off of school_data_complete
small_school_data = school_data_complete.loc[:,["school_name", "type", "size", "budget", "reading_score",
                                    "math_score"]]

# calculate the values for the index in the dataframe
small_total_students = small_school_data.groupby("school_name")["size"].first()
small_total_budget = small_school_data.groupby("school_name")["budget"].first()
small_per_student_budget = small_total_budget/ small_total_students
small_avg_math_score = small_school_data.groupby("school_name")["math_score"].mean()
small_avg_reading_score = small_school_data.groupby("school_name")["reading_score"].mean()
small_math_70 = (math_scores_70_df.groupby("school_name")["math_score"].count()/small_total_students)*100
small_reading_70 = (reading_scores_70_df.groupby("school_name")["reading_score"].count()/small_total_students)*100
small_overall_pass = (small_math_70 + small_reading_70)/2

# stored data from groupby into a dataframe
small_school_summary_df = pd.DataFrame({" ": small_school_data.groupby("school_name")["school_name"].first(),
                            "School Type": small_school_data.groupby("school_name")["type"].first(), 
                            "Total Students": small_total_students, "Total School Budget":small_total_budget,
                            "Per Student Budget": small_per_student_budget,
                            "Average Math Score": small_avg_math_score, "Average Reading Score": small_avg_reading_score,
                            "% Passing Math": small_math_70, "% Passing Reading":small_reading_70,
                            "Overall Passing Rate": small_overall_pass
})

# change the index
small_school_summary_df.set_index(keys=" ", inplace=True)


# format the budget to refelct currency
small_school_summary_df["Total School Budget"] = small_school_summary_df["Total School Budget"].map("${:,.2f}".format)
small_school_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# group the data by grade and calculate average math score
math_9 = school_data_complete.loc[school_data_complete["grade"] == "9th"]
group_math_9 = math_9.groupby("school_name")["math_score"].sum() / math_9.groupby("school_name")["Student ID"].count()

math_10 = school_data_complete.loc[school_data_complete["grade"] == "10th"]
group_math_10 = math_10.groupby("school_name")["math_score"].sum() / math_10.groupby("school_name")["Student ID"].count()

math_11 = school_data_complete.loc[school_data_complete["grade"] == "11th"]
group_math_11 = math_10.groupby("school_name")["math_score"].sum() / math_11.groupby("school_name")["Student ID"].count()

math_12 = school_data_complete.loc[school_data_complete["grade"] == "10th"]
group_math_12 = math_12.groupby("school_name")["math_score"].sum() / math_12.groupby("school_name")["Student ID"].count()

math_df = pd.DataFrame({" ": small_school_data.groupby("school_name")["school_name"].first(),
                            "School Type": small_school_data.groupby("school_name")["type"].first(), 
                            "9th Grade": group_math_9, "10th Grade":group_math_10,
                            "11th Grade": group_math_11,
                            "12th Grade": group_math_12})

# change the index
math_df.set_index(keys=" ", inplace=True)

# Format the grades
math_df["9th Grade"] = math_df["9th Grade"].map("{:.2f}".format)
math_df["10th Grade"] = math_df["10th Grade"].map("{:.2f}".format)
math_df["11th Grade"] = math_df["11th Grade"].map("{:.2f}".format)
math_df["12th Grade"] = math_df["12th Grade"].map("{:.2f}".format)
math_df = math_df.sort_values("9th Grade", ascending = 0)
math_df
```
